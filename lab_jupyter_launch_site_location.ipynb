```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Import necessary libraries
from geopy.distance import geodesic

# Define a function to calculate the distance between two points (latitude, longitude)
def calculate_distance(coord1, coord2):
    """
    Calculate the distance between two coordinates using geopy.
    coord1, coord2: tuples of (latitude, longitude)
    """
    return geodesic(coord1, coord2).km

# Sample proximity points; replace with actual points of interest
# Example: Nearby cities or landmarks
proximities = {
    'NASA Johnson Space Center': (29.559684888503615, -95.0830971930759),
    'Orlando': (28.5383, -81.3792),
    'Los Angeles': (34.0522, -118.2437)
}

# Create a new column in the dataframe to store distances to each proximity point
distance_results = []

# Loop through each launch site and calculate distances to each proximity
for index, row in launch_sites_df.iterrows():
    site_coord = (row['Lat'], row['Long'])
    site_distances = {'Launch Site': row['Launch Site']}
    
    # Calculate the distance from the launch site to each proximity point
    for location, coord in proximities.items():
        distance = calculate_distance(site_coord, coord)
        site_distances[f'Distance to {location} (km)'] = distance
    
    distance_results.append(site_distances)

# Convert the results into a DataFrame for better visualization
distance_df = pd.DataFrame(distance_results)
distance_df.head()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordinates of the launch site (example for CCAFS SLC-40, adjust as necessary)
launch_site_lat, launch_site_lon = 28.563197, -80.576820

# Example coordinates for other points of interest - replace with actual values found using MousePosition
closest_city_coords = (28.6123, -81.2034)  # Replace with actual coordinates of the closest city
closest_railway_coords = (28.5721, -80.5857)  # Replace with coordinates of the closest railway
closest_highway_coords = (28.5739, -80.6550)  # Replace with coordinates of the closest highway

# Calculate distances to each point of interest
distance_to_city = calculate_distance(launch_site_lat, launch_site_lon, closest_city_coords[0], closest_city_coords[1])
distance_to_railway = calculate_distance(launch_site_lat, launch_site_lon, closest_railway_coords[0], closest_railway_coords[1])
distance_to_highway = calculate_distance(launch_site_lat, launch_site_lon, closest_highway_coords[0], closest_highway_coords[1])

# Function to create markers and lines on the map
def add_marker_and_line(coord, label, distance):
    # Add a marker with the distance to the point of interest
    marker = folium.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{label}: {distance:.2f} KM</b></div>'
        )
    )
    # Add the marker to the map
    site_map.add_child(marker)

    # Draw a line between the launch site and the point of interest
    line = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], coord], weight=2, color='blue')
    site_map.add_child(line)

# Add markers and lines for city, railway, and highway
add_marker_and_line(closest_city_coords, 'City', distance_to_city)
add_marker_and_line(closest_railway_coords, 'Railway', distance_to_railway)
add_marker_and_line(closest_highway_coords, 'Highway', distance_to_highway)

# Display the map
site_map
```
